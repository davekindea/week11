```python
import yfinance as yf

# Define the tickers and the date range
tickers = ['TSLA', 'BND', 'SPY']
start_date = '2015-01-01'
end_date = '2024-10-31'

# Fetch historical data using yfinance
data = {ticker: yf.download(ticker, start=start_date, end=end_date) for ticker in tickers}

# Save each DataFrame as a CSV file
for ticker, df in data.items():
    df.to_csv(f'{ticker}_historical_data.csv')  # Save to CSV with ticker name
    print(f'Saved {ticker} data to CSV')
```

```python
for ticker in tickers:
    data[ticker].to_csv(f"{ticker}_historical_data.csv")
```

```python
data_TSLA['Rolling Mean (20 days)'] = data_TSLA['Close'].rolling(window=20).mean()
data_TSLA['Rolling Std Dev (20 days)'] = data_TSLA['Close'].rolling(window=20).std()

# Plot the closing price along with the rolling mean and standard deviation
plt.figure(figsize=(10,6))
plt.plot(data_TSLA['Close'], label='TSLA Closing Price')
plt.plot(data_TSLA['Rolling Mean (20 days)'], label='20-Day Rolling Mean', linestyle='--')
plt.plot(data_TSLA['Rolling Std Dev (20 days)'], label='20-Day Rolling Std Dev', linestyle=':')
plt.title('TSLA Closing Price with 20-Day Rolling Mean and Std Dev')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Calculate the Z-score for the closing price
data_TSLA['Z-Score'] = (data_TSLA['Close'] - data_TSLA['Close'].mean()) / data_TSLA['Close'].std()

# Find outliers (Z-score > 3 or Z-score < -3)
outliers = data_TSLA[data_TSLA['Z-Score'].abs() > 3]

# Plot the outliers
plt.figure(figsize=(10,6))
plt.plot(data_TSLA['Close'], label='TSLA Closing Price')
plt.scatter(outliers.index, outliers['Close'], color='red', label='Outliers')
plt.title('TSLA Outliers Based on Z-Score')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

print(outliers[['Z-Score', 'Close']])
```

```python
# Find days with unusually high or low returns
unusual_returns = data_TSLA[(data_TSLA['Daily Change %'] > 5) | (data_TSLA['Daily Change %'] < -5)]

# Plot the returns along with the unusually high/low return days
plt.figure(figsize=(10,6))
plt.plot(data_TSLA['Daily Change %'], label='TSLA Daily Change (%)', color='orange')
plt.scatter(unusual_returns.index, unusual_returns['Daily Change %'], color='red', label='Unusual Returns')
plt.title('TSLA Unusual Daily Returns (Greater than 5% or less than -5%)')
plt.xlabel('Date')
plt.ylabel('Daily Change (%)')
plt.legend()
plt.grid(True)
plt.show()

print(unusual_returns[['Daily Change %', 'Close']])
```
